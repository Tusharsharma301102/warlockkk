```python
#ans3
def count_occurrences(string, word):
  count = 0
  for i in range(len(string)):
    if string[i:i + len(word)] == word:
      count += 1
  return count



string = input()
count_cat = count_occurrences(string, "cat")
count_dog = count_occurrences(string, "dog")
if count_cat==count_dog:
  print("true")
else:
    print("false")
```

```python
#ans4
import calendar

def print_season(month, day):
  seasons = {
    1: "Winter",
    2: "Winter",
    3: "Spring",
    4: "Spring",
    5: "Spring",
    6: "Summer",
    7: "Summer",
    8: "Summer",
    9: "Fall",
    10: "Fall",
    11: "Fall",
    12: "Winter"
  }
  return seasons[month]

def is_leap_year(year):
  if year % 4 == 0:
    if year % 100 == 0:
      if year % 400 == 0:
        return True
      else:
        return False
    else:
      return True
  else:
    return False

def print_leap_year(year):
  next_leap_year = year + 1
  while not is_leap_year(next_leap_year):
    next_leap_year += 1
  return next_leap_year


def main():
  year = int(input("Enter a year: "))
  month = int(input("Enter a month: "))
  day = int(input("Enter a day: "))
  season = print_season(month, day)
  print("The season of", month, day, "is", season)
  if is_leap_year(year):
    print("The number of days in", year, "is", calendar.leapdays(year, year + 1))
  else:
    print("The next leap year after", year, "is", print_leap_year(year))


if __name__ == "_main_":
  main()
```

```python
#ans6
def even_filter(list_of_numbers):
  even_numbers = []
  for number in list_of_numbers:
    if number % 2 == 0:
      even_numbers.append(number)
  return even_numbers


def odd_filter(list_of_numbers):
  odd_numbers = []
  for number in list_of_numbers:
    if number % 2 == 1:
      odd_numbers.append(number)
  return odd_numbers


def main():
  list_of_numbers = []
  print("Enter a list of numbers: ")
  for number in input().split():
    list_of_numbers.append(int(number))

  even_numbers = even_filter(list_of_numbers)
  print("The even numbers are: ", even_numbers)

  odd_numbers = odd_filter(list_of_numbers)
  print("The odd numbers are: ", odd_numbers)


if __name__ == "_main_":
  main()
```

```python
#ans7
def gcd(x, y):
  if x == 0:
    return y
  else:
    return gcd(y % x, x)


def main():
  list_of_numbers = []
  print("Enter a list of numbers: ")
  for number in input().split():
    list_of_numbers.append(int(number))

  gcd_of_numbers = gcd(list_of_numbers[0], list_of_numbers[1])
  for i in range(2, len(list_of_numbers)):
    gcd_of_numbers = gcd(gcd_of_numbers, list_of_numbers[i])

  print("The GCD of the numbers is: ", gcd_of_numbers)


if __name__ == "_main_":
  main()
```

```python
#ans8
def check_prime(number):
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True


def twin_prime(number):
  twin_primes = []
  for i in range(3, number):
    if check_prime(i):
      if check_prime(i + 2):
        twin_primes.append((i, i + 2))
  return twin_primes


def main():
  twin_primes = twin_prime(1000)
  print("The twin primes less than 1000 are: ", twin_primes)


if __name__ == "_main_":
  main()
```

```python
#ans9
import math

def permutations(n, r):
  if r > n:
    raise ValueError("r cannot be greater than n")
  return math.factorial(n) // math.factorial(n - r)


def combinations(n, r):
  return permutations(n, r) // math.factorial(r)


def main():
  print("The number of permutations of 5 objects taken 2 at a time is: ",
        permutations(5, 2))
  print("The number of combinations of 5 objects taken 2 at a time is: ",
        combinations(5, 2))


if __name__ == "_main_":
  main()
```

```python
#ans10
def prodDigits(number):
  product = 1
  for digit in str(number):
    product *= int(digit)
  return product


def MDR(number):
  product = prodDigits(number)
  while len(str(product)) > 1:
    product = prodDigits(product)
  return product


def MPersistence(number):
  persistence = 0
  product = prodDigits(number)
  while len(str(product)) > 1:
    persistence += 1
    product = prodDigits(product)
  return persistence


def main():
  number = 86
  print("The multiplicative digital root of ", number, " is: ", MDR(number))
  print("The multiplicative persistence of ", number, " is: ", MPersistence(number))


if __name__ == "_main_":
  main()
```

```python
#ans12
#The NAND operation is true when at least one of its operands is false
#we can use the NOT (not) operator along with the OR (or) operator to achieve the NAND functionality.
def nand(A, B):
    return not (A and B)

# Example usage
A = True
B = False
result = nand(A, B)
print(result)  # Output: True
```
